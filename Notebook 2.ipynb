```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['cnt'], kde=True)
plt.title('Distribution of Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['temp'], y=df['cnt'])
plt.title('Relationship between Temperature and Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['hum'], y=df['cnt'])
plt.title('Relationship between Humidity and Count')
plt.show()
```

```python
df=pd.read_csv('final_merged_output.csv')
df.head(1000)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['cnt'], kde=True)
plt.title('Distribution of Bike Rentals')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['season'], y=df['cnt'])
plt.title('Seasonal Trends in Bike Rentals')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['weathersit'], y=df['cnt'])
plt.title('Weather Impact on Bike Rentals')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df['hr'], y=df['cnt'])
plt.title('Hourly Trends in Bike Rentals')
plt.show()
```
